```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import numpy as np
import random
import warnings
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(rc={'figure.figsize':(13,7.5)})
sns.set_context('talk')
```

```python
#calculate population mean
pop_mean_borough = df['BOROUGH'].mean()
pop_mean_borough
```

```python
#calculate population std
pop_std_borough = df['BOROUGH'].std()
pop_std_borough
```

```python
# mengambil 300 sampel masing" dr 25 nilai acak
sample_means = []
n = 25
for sample in range(0,300):
    #random sampling done with replacement
    sample_values = np.random.choice(a=df['BOROUGH'], size=n)
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
# view distribution pf sample means (otherwise called sampling distribution)
sns.distplot(sample_means)
plt.title("Distribution of sample means ($n=25$) of BOROUGH", y=1.015, fontsize=20)
plt.xlabel('sample BOROUGH', labelpad=14)
plt.ylabel('frequency of occurence', labelpad=14)
```

berdasarkan plooting Distribution of sample means (n=25) of BOROUGH dapat dilihat bahwa kurva cukup simetris sesuai distribusinya

```python
# calculate sapling distribution (n=25) summary statistics

#calculate median of sample means
median_of_sample_means = np.median(sample_means)
median_of_sample_means
```

```python
# calculate mean of sample means
mean_of_sample_means = np.mean(sample_means)
mean_of_sample_means
```

```python
pop_mean_borough

# nilai mean_of_sample_means kira" sama dgn nilai population mean yg ditetapkan ke variabel pop_mean_mass
```

```python
# calculate std of sample means
std_dev_of_sample_means = np.std(sample_means)
std_dev_of_sample_means
```

```python
# Equation for std of sampling distribution
standard_error = pop_std_borough/np.sqrt(n)
standard_error
```

```python
std_dev_of_sample_means
```

```python
# Critcal Z-Scores
tail = 1 - stats.norm.cdf(1.96)
tail
```

```python
# hitung batas yg tepat
lower_95_perc_bound = mean_of_sample_means - 1.96*standard_error
lower_95_perc_bound
```

```python
upper_95_perc_bound = mean_of_sample_means + 1.96*standard_error
upper_95_perc_bound
```

## 11. Hypothesis Testing
Membuat 1 hipotesis dan menarik kesimpulan yang ada

**Research Question:** apakah NEIGBORHOOD ALPHABET CITY memiliki SALE PRICE rata-rata yang lebih tinggi secara signifikan daripada WOODROW?

H0 : rata-rata SALE PRICE NEIGBORHOOD ALPHABET CITY lebih tinggi SALE PRICE NEIGBORHOOD WOODROW

H1 : SALE PRICE NEIGBORHOOD ALPHABET CITY kurang dari SALE PRICE NEIGBORHOOD WOODROW

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_Neigh = df[["NEIGHBORHOOD","SALE PRICE"]]
df_alpha= df_Neigh[df_Neigh['NEIGHBORHOOD'] == 'ALPHABET CITY']
df_alpha.head()
```

```python
df_wood = df_Neigh[df_Neigh['NEIGHBORHOOD'] == 'WOODROW']
df_wood.head()
```

```python
n1 = len(df_alpha)
mu1 = df_alpha["SALE PRICE"].mean()
sd1 = df_alpha["SALE PRICE"].std()

(n1, mu1, sd1)
```

```python
n2 = len(df_wood)
mu2 = df_wood["SALE PRICE"].mean()
sd2 = df_wood["SALE PRICE"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(df_alpha["SALE PRICE"].dropna(), df_wood["SALE PRICE"].dropna(),alternative='two-sided')
```

```python
plt.figure(figsize=(7,4))
plt.title('SALE PRICE NEIGBORHOOD ALPHABET CITY', fontsize=16)
plt.hist(df_alpha["SALE PRICE"].dropna(),edgecolor='k', color='green',bins=25)
plt.show()

plt.figure(figsize=(7,4))
plt.title('SALE PRICE NEIGBORHOOD WOODROW', fontsize=16)
plt.hist(df_wood["SALE PRICE"].dropna(),edgecolor='k', color='blue',bins=25)
plt.show()
```

**Kesimpulan:** karena p-value (0.006181179411764059) sgt kecil, kita dapat reject the null hypothesis atau menerima H0 bahwa rata-rata SALE PRICE NEIGBORHOOD ALPHABET CITY lebih tinggi secara signifikan daripada SALE PRICE NEIGBORHOOD WOODROW. catatan, kita menggunakan alternative='two-sided' dlm z-test krn disini kita memeriksa ketidaksamaan.
