## Maps with markers

```python
df_incidents = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')

print('Dataset downloaded & read into a pandas dataframe')
```

```python
#melihat 5 item pertama dataset
df_incidents.head()
```

```python
df_incidents.shape
```

```python
#get the first 100 crimes in the df_incidents dataframe
limit=100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
df_incidents.shape
```

visualisasi dmana kejahatan terjai di kota san fransisco. kita akan menggunakan gaya default & kita akan menginisialisasi tingkat zoom ke 12

```python
#san fransisco latitude & longtitude values
latitude=37.77
longtitude=-122.2
```

```python
#create map & display it
sanfran_map = folium.Map(location=[latitude, longtitude], zoom_start=12)

sanfran_map
```

sekarang mari taruh lokasi kejahatan ke dlm peta dgn membuat fature group dgn fitur & gayanya sendiri, lalu menambahkannya ke sanfran_map

```python
#instantiate a feature group for the inccidents in the dataframe
incidents = folium.map.FeatureGroup()

#loop through the 100 crimes and add each to the incidents feature group
for  lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng], 
            radius=5, #define how big u want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
    
#add incidents to map
sanfran_map.add_child(incidents)
```

kita juga dpt menambahkan beberapa teks pop up yg akan ditampilkan saat kalian mengarahkan kursor ke atas marker. mari buat setiap penanda menampilkan kategori kejahatan saat diarahkan ke atas marker

```python
#instantiate a feature group for the incidents in  the dataframe
incidents = folium.map.FeatureGroup()

#loop through the 100 crimes and add each to the incidents feature group
for  lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng], 
            radius=5, #define how big u want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

#add pop up text to each marker in the map
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)
    
#add incidents to map
sanfran_map.add_child(incidents)
```

sekarang kiita dpt mengetahui kategori kejahatan apa yg terjadi di setiap marker. jika kita menyadari peta menjadi sgt padat akan semua marker ini, ada 2 solusi utk masalah ini. solusi yg lbh sederhana adalah dgn menghapus marker lokasi ini & menambahkan teks ke marker lingkaran itu senderi sbg berikut:

```python
#create map & display it
sanfran_map = folium.Map(location=[latitude, longtitude], zoom_start=12)

#loop through the 100 crimes and add each to map
for  lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.CircleMarker(
        [lat, lng], 
        radius=5, #define how big u want the circle markers to be
        color='yellow',
        fill=True,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sanfran_map)
        
#show map
sanfran_map
```

solusi lain yg tepay yaitu dgn mengelompokkan marker ke dlm cluster yg berbeda. setiap cluster kemudian diwakili oleh jumlah kejahatan di setiap lingkungan. cluster ini dianggap sbg kantong dr kejahatan san francisco yg kemudian dpt kita analisis scr terpisah

kita mulai dgn membuat instance objek markercluster & menambahkan semua titik data si dataframe ke objek ini

```python
from folium import plugins

#lets start again with a cclean copy of the map of san francisco
sanfran_map = folium.Map(location=[latitude, longtitude], zoom_start=12)

#instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(sanfran_map)

#loop through the 100 crimes and add each to map
for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Category):
    folium.Marker(
        location=[lat, lng], 
        icon=None,
        popup=label,
    ).add_to(incidents)
        
#display map
sanfran_map
```

## Choropleth Maps
yaitu peta tematik dimana area diarsir / berpola sebanding dgn pengukuran statistik yg ditampilkan di peta, seperti kepadatan penduduk/pendapatan per kapita. ini cara mudah utk memvisualisasikan bagaimana suatu pengukuran bervariasi di seluruh wilayah geografis/menunjukkan tingkat variabilitas dlm suatu wilayah.

```python
#download dataset & tampilkan kedalam pandas dataframe
df_can = pd.read_excel('Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

print('Data downloaded and read into a dataframe')
```

```python
#melihat 5 item pertama dataset
df_can.head()
```

```python
df_cann = df_can.dropna(axis=1)
```

```python
df_cann
```

```python
#banyaknya entri dataset
print(df_cann.shape)
```

```python
#clean up the dataset to remove unnecesssary columns (eg. REG)
df_cann.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

#lets rename the columns so that they make sense
df_cann.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)

#for sake of consistency, lets also make all column labels of type string
df_cann.columns = list(map(str, df_cann.columns))

#add total column
df_cann['Total'] = df_cann.sum(axis=1)

#years that we will be using in this lession - useful for plotting later on
years=list(map(str, range(1980, 2014)))
print('data dimension:', df_cann.shape)
```

```python
df_cann.head()
```

untuk membuat peta choropleth, kita membutuhkan file GeoJSON yg mendefinisikan are / batas negara bagian, kabupaten, atau negara yg kita minati. dalam kasus ini, krn kita berusaha utk membuat peta dunia, kita menginginkan sebuah GeoJSON yg mendefinisikan batas" semua negara lain

mari kita buat peta dunia, berpusat di sekitar [0,0] nilai lintang, dan bujur, dengan tingkat zoom awal , dan menggunakan gaya Mapbox Bright

```python
df_can.reset_index(inplace=True)
```

```python
world_geo = r'world_countries.json' #geojson file

#create a plain world map
world_map = folium.Map(location=[0,0], zoom_start=2, tiles="Stamen Terrain")

#generate choropleth map using the total immigration of each ountry to canada from 1980 - 2013
world_map.choropleth(
    geo_data=world_geo,
    data=df_cann,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', #yellow, orange, red
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Immigration to canada'
)

#display map
world_map
```
