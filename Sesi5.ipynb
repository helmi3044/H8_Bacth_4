## Getting to Know Pandas' Data Structures
### Understanding Series Object
List merupakan struktur data python paling dasar dan awal baik untuk mengenal objek panda series

```python
#membuat objek series baru menggunakan list
revenues= pd.Series([5555,7000,1980])
```

objeks series mempunyai 2 komponen:
1. a sequence of values
2. a sequence of identifiers, which is the index

kita dapat mengakses komponen ini masing" dengan .values  dan .index :

```python
revenues
```

revenues.values mengembalikan nilai dalam seri, sedangkan revenues.index mengembalikan indeks posisi

```python
revenues.values
```

```python
revenues.index
```

pandas series memiliki indeks integer yg ditentukan scr implisit. indeks implisit ini menunjukkan posisi elemen dalam series. Namun, series juga dapat memiliki jenis indeks yang berubah-ubah. Kalian dapat menganggap indeks eksplisit ini sebagai label untuk baris tertentu:

```python
city_revenues = pd.Series(
    [4200, 8000, 6500],
    index=["Amsterdam", "Toronto", "Tokyo"]
)
city_revenues
```

INDEX : DAFTAR NAMA KOTA STRING

- revenues : this series behaves like a python list because it only has a positional index
- city_revenue : this series acts like a python dictionary because it features both a positional and a label index

berikut cara membuat series dengan label dari dictionary python:

```python
city_employee_count = pd.Series({"Amsterdam":5, "Tokyo":8})
city_employee_count
```

dictionary keys menjadi index, dan nilai dictionary adalah nilai series.

series juga mendukung .keys dan keyword in:

```python
city_employee_count.keys()
```

<b>Using .loc and .iloc</b>

indek([]) memang mudah tetapi bagaimana jika labelnya juga angka?

```python
colors = pd.Series(
    ["red", "purple", "blue", "green", "yellow"],
    index=[1,2,3,5,8]
)

colors
```

jika kta menggunakan indeks label maka colors[1] menacu pada "red"

untuk meghindari kita bisa gunakan:
1. .loc refers to the label inde
2. .iloc refers to the positional index

```python
colors.loc[1] #elemen dgn label 1
```

```python
colors.iloc[1] #elemen dgn indeks 1
```

.loc & .iloc juga mendukung fitur slicing. namun, metode akses data ini memiliki perbedaan penting. sementara .iloc mengecualikan elemen penutup, .iloc menyertakannya.

```python
#return the elements with the implicit index: 1, 2
colors.iloc[1:3]
```

disini iloc menutup itm "hijau" dengan indeks 3 dikecualikan. disisi lain .loc menyertakan elemen penutup

```python
#return the elements with the explicit index between 3 and 8
colors.loc[3:8]
```

```python
#kita juga bisa menyertakan indeks posisi negatif ke .iloc
colors.iloc[-2]
```

kita dpt menggunakan blok kode di atas utk membedakan antara 2 perilaku series:
- you can use .iloc on a series similar using [] on list
- you can use .loc on a seriess similar to using [] on a dictionary

### Combining Multiple Dataset
disini menggunakan .concat()

```python
further_city_data = pd.DataFrame(
    {"revenue": [7000,3400], "employee_count": [2,2]},
    index=["New York", "Barcelona"]
)
further_city_data
```

```python
city_data
```

```python
all_city_data = pd.concat([city_data, further_city_data], sort=False)
all_city_data
```

scr defaut .concat() menggabungkan sepanjang axis=0. artinya cconcat menambahkan baris. kita dapat menggunakannya utk menambahkan kolom dgn menyediakan parameter axis=1

```python
city_countries = pd.DataFrame({
    "country": ["Holland", "Japan", "Holland", "Canada", "Spain"],
    "capital": [1,1,0,0,0]},
    index=["Amsterdam", "Tokyo", "Rotterdam", "Toronto", "Barcelona"]
)

city_countries
```

```python
all_city_data
```

```python
cities = pd.concat([all_city_data, city_countries], axis=1, sort=False)
cities
```

```python
#jika menggabungkan hanya kota yg muncul di kedua objek DataFrame, kita dpt mengatur parmeter join menjadi inner
pd.concat([all_city_data, city_countries], axis=1, join="inner")
```

kita bisa menggunakan .merge() utk mengimplementasikan operasi gabungan yg mirip dgn SQL:

```python
countries = pd.DataFrame({
    "population_millions": [17,127,37],
    "continent": ["Europe", "Asia", "North America"]},
    index=["Holland", "Japan", "Canada"]
)
```

```python
countries
```

```python
cities
```

```python
pd.merge(cities, countries, left_on="country", right_index=True)
```

.merge() melakukan inner join scr default. Jika kita ingin memasukkan semua kota dlm hasil, maka kita perlu memberikan parameter how:

```python
pd.merge(
    cities,
    countries,
    left_on="country",
    right_index=True,
    how="left" # dgn ini kita akan melihat semua kota, termasuk yg tdk memiliki ata negara
)
```
